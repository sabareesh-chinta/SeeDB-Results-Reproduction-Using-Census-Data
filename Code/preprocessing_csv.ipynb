```python
data_file = "../Data/ev_population.csv"
df = pd.read_csv(data_file)
print(df)
```

```python
train_df, test_df = train_test_split(df, test_size=0.67, random_state=42)
```

```python
full_df = pd.concat([train_df, test_df])
full_df.info()
```

```python
us = ["CADILLAC", "CHEVROLET", "CHRYSLER", "FISKER", "FORD", "JEEP", "LINCOLN", "LUCID", "RIVIAN", "TESLA", "WHEEGO ELECTRIC CARS"]
```

```python
full_df['Brand Origin'] = ['US' if make in us else 'Rest of the World' for make in full_df['Make']]
full_df.head()
```

```python
full_df.to_csv("../Data/ev_population_total.csv", index=False)
```
